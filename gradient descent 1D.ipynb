```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=10*np.random.rand(30)
x
```

```python
y=12*x +3
```

```python
plt.scatter(x,y)
```

```python
def cost(y,x,m,c):
    ans=0
    for i in range(len(x)):
        ans+=(y[i]-m*x[i]-c)**2
    return ans
def slope(x,y,m,c):
    N=len(x)
    dm=0
    dc=0
    for i in range(N):
        dm+=(-2/N)*(y[i]-m*x[i]-c)*x[i]
        dc+=(-2/N)*(y[i]-m*x[i]-c)
    return dm,dc
def grad_desc(x,y,alpha,it):
    m=0
    c=0
    for i in range(it):
        dm,dc=slope(x,y,m,c)
        m-=alpha*dm
        c-=alpha*dc
        print(cost(y,x,m,c))
    
    return m,c
        
```

```python
alpha=0.001
it=1000
m,c=grad_desc(x,y,alpha,it)
```

```python
m,c
```

```python
pred=m*x +c
```

```python
plt.scatter(x,y)
plt.plot(x,pred,'r')
```
