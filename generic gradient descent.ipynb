```python
x=np.array(data.data)
y=np.array(data.target)
```

```python
x=np.insert(x,np.shape(x)[1],[1]*len(x),axis=1)
```

```python
def slope(x,y,m,N,M):
    ans=np.zeros(N)
    for i in range(M):
        for j in range(N):
            ans[j]+=(-2/M)*(y[i]-(m*x[i]).sum())*x[i][j]
    return ans

def cost(x,y,m,M):
    ans=0
    for i in range(M):
        ans+=(y[i]-(m*x[i]).sum())**2
    return ans
def grad(x,y,alpha,it):
    N=np.shape(x)[1]
    M=np.shape(x)[0]
    m=np.zeros(N)
    for i in range(it):
        dx=slope(x,y,m,N,M)
        for j in range(N):
            m[j]-=alpha*dx[j]
        print(cost(x,y,m,M))
    return m
```

```python
alpha=0.000003
it=500
m=grad(x,y,alpha,it)
```
